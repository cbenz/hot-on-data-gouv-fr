```python
import csv
import datetime
import json
import requests
```

```python
CSV_FILE_PATH = u'datasets_data.csv'
DATE = datetime.date(2014, 4, 4)
END_DATE = datetime.date(2014, 4, 14)
HOST_NAME = u'stats.data.gouv.fr'
```

```python
def fetch_page_urls_data(date, end_date = None, subtable_id = None):
    url = u'http://{host_name}/index.php'.format(host_name = HOST_NAME)
    date_str = unicode(date)
    if end_date is not None:
        date_str += u',' + unicode(end_date)
    params = {
        'date': date_str,
        'filter_limit': 100,
        'filter_sort_column': u'nb_visits',
        'filter_sort_order': u'desc',
        'flat': 0,
        'format': u'JSON',
        'idSite': 1,
        'method': u'Actions.getPageUrls',
        'module': u'API',
        'period': u'day',
        'token_auth': u'anonymous',
        }
    if subtable_id is not None:
        params['idSubtable'] = subtable_id
    response = requests.get(url, params = params)
    return response.json()
```

```python
page_urls_data = fetch_page_urls_data(date = DATE, end_date = END_DATE)
```

```python
print json.dumps(page_urls_data, indent = 2)
```

```python
find_subtable_id = lambda urls_data, label: [url_data for url_data in urls_data if url_data['label'] == label][0]['idsubdatatable']

def fetch_datasets_data(date, end_date = None):
    data = fetch_page_urls_data(date = date, end_date = end_date)
    for label in (u'fr', u'dataset'):
        data = {
            date_str: fetch_page_urls_data(
                date = date_str,
                subtable_id = find_subtable_id(urls_data, label),
                )
            for date_str, urls_data in data.iteritems()
            }
    return data
```

```python
datasets_data = fetch_datasets_data(date = DATE, end_date = END_DATE)
```

```python
print json.dumps(datasets_data, indent = 2)
```

```python
def datasets_data_to_csv_rows(datasets_data):
    keys = ['url', 'label', 'sum_time_spent', 'nb_uniq_visitors', 'bounce_rate', 'nb_hits_following_search', 'exit_rate',
            'entry_sum_visit_length', 'avg_time_on_page', 'entry_nb_uniq_visitors', 'sum_time_spent',
            'exit_nb_visits', 'max_time_generation']
    rows = []
    for date_str, urls_data in datasets_data.iteritems():
        for url_data in urls_data:
            if url_data['label'].startswith('/') and url_data['label'] not in [u'/new', u'/index']:
                row = {key: unicode(url_data.get(key, u'')).encode('utf-8') for key in keys}
                row['date'] = u'/'.join([date_str[8:11], date_str[5:7], date_str[0:4]])
                rows.append(row)
    return sorted(rows, key = lambda item: (item['url'], item['date'], item['nb_uniq_visitors']))
```

```python
csv_rows = datasets_data_to_csv_rows(datasets_data)
```

```python
print json.dumps(datasets_data, indent = 2)
```

```python
fieldnames = ['date', 'url', 'label', 'sum_time_spent', 'nb_uniq_visitors', 'bounce_rate', 'nb_hits_following_search', 'exit_rate',
             'entry_sum_visit_length', 'avg_time_on_page', 'entry_nb_uniq_visitors', 'sum_time_spent',
             'exit_nb_visits', 'max_time_generation']
with open(CSV_FILE_PATH, 'w') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames = fieldnames)
    writer.writeheader()
    writer.writerows(csv_rows)
```
